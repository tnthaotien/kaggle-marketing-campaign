```python
df.Total_Purchases.describe()
```

```python
def purchases(Total_Purchases):
  if Total_Purchases < 8:
    return ("Low")
  elif 8 <= Total_Purchases <= 15:
    return ("Medium")
  else:
    return ("High")
```

```python
df['Purchase_Segment'] = df['Total_Purchases'].apply(lambda x: purchases(x))
```

```python
def expense(Sum_Mnt_Prods):
  if Sum_Mnt_Prods < 60:
    return ("Low")
  elif 60 <= Sum_Mnt_Prods <= 350:
    return ("Medium")
  else:
    return ("High")
```

```python
df['Mnt_Segment'] = df['Sum_Mnt_Prods'].apply(lambda x: expense(x))
```
